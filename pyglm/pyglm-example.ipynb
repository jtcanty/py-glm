In this notebook I will be performing regression using `pyglm.py` with simulated data on different error distributions in the exponential family of GLMs
### Import libraries

```python
import numpy as np
import random
import pyglm as glm
import matplotlib.pyplot as plt
```

### Gaussian Distribution

Let $Y \sim N(\mu,\sigma^2)$, where the canonical link function is defined as $ \mu = X\beta$

```python
# Simulate a line with Gaussian noise
m = 0.5
b = 1
x1 = 1 * np.linspace(1,100,100)

noise = np.zeros(100)
for i in range(0,100):
    noise[i] = random.normalvariate(0,2)

y1 = m * x1 + b + noise
```

```python
plt.scatter(x1, y1, marker='.', color='b')
plt.show()
```

```python
# Fit using py-glm

# Newton's method
clf = glm.GLM(distr='gaussian', n_iter=100, descent='N', learning_rate = 10)
n_param = clf.fit(x1 ,y1)

# Batch gradient descent
clf = glm.GLM(distr='gaussian', n_iter=100, descent='BGD', learning_rate = .001)
b_param = clf.fit(x1 ,y1)

# Stochastic gradient descent
clf = glm.GLM(distr='gaussian', n_iter=1000, descent='SGD', learning_rate = .0001)
s_param = clf.fit(x1 ,y1)

```

```python
# Plot fit

plt.scatter(x1, y1, marker='.', color='b')
plt.plot(x1, n_param[1] * x1 + n_param[0], color='g', label='Newton')
plt.plot(x1, b_param[1] * x1 + b_param[0], color='r', label='Batch gradient')
plt.plot(x1, s_param[1] * x1 + s_param[0], color='k', label='Stochastic gradient')
plt.title('Fit Result')
plt.legend()

plt.show()
```

### Poisson Distribution

Let $Y \sim Poisson(\lambda)$, where the canonical link function is defined as $ \log(\mu) = X\beta$

```python
# Simulate data with Poisson error distribution

x2 = np.linspace(0, 10, 100)

beta0 = 1
beta1 = 0.5

mu = np.exp(beta0 + beta1 * x2)
y2 = np.random.poisson(mu)
```

```python
plt.scatter(x2, y2)
plt.show()
```

```python
# Fit using py-glm

# Newton's method
clf = glm.GLM(distr='poisson', n_iter=100, descent='N', learning_rate = 1)
n_param = clf.fit(x2 ,y2)

# Batch gradient descent
clf = glm.GLM(distr='poisson', n_iter=1000, descent='BGD', learning_rate = .0001)
b_param = clf.fit(x2 ,y2)

# Stochastic gradient descent
clf = glm.GLM(distr='poisson', n_iter=2000, descent='SGD', learning_rate = .00001)
s_param = clf.fit(x2 ,y2)
```

```python
# Plot fit

plt.scatter(x2, y2, marker='.', color='b')
plt.plot(x2, np.exp(n_param[1] * x2 + n_param[0]), color='g', label='Newton')
plt.plot(x2, np.exp(b_param[1] * x2 + b_param[0]), color='r', label='Batch gradient')
plt.plot(x2, np.exp(s_param[1] * x2 + s_param[0]), color='k', label='Stochastic gradient')
plt.title('Fit Result')
plt.legend()

plt.show()
```

### Bernoulli Distribution

Let $Y \sim Bernoulli(p)$, where the canonical link function is defined as $ \log(\frac{\mu}{(1 - \mu)}) = X\beta$

```python
x = np.random.randn(100)
y = np.zeros(100)

z = 1 + 2 * x
pr = 1 / (1 + np.exp(-z))

for i in range(0, len(z)):
    y[i] = np.random.binomial(1, pr[i])
```

```python
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('Probability')

plt.show()
```

```python
# Fit using py-glm

# Newton's method
clf = glm.GLM(distr='categorical', n_iter=1000, descent='N', learning_rate = 10)
n_param = clf.fit(x ,y)

# Batch gradient descent
clf = glm.GLM(distr='categorical', n_iter=1000, descent='BGD', learning_rate = 3.8)
b_param = clf.fit(x ,y)

# Stochastic gradient descent
clf = glm.GLM(distr='categorical', n_iter=1000, descent='SGD', learning_rate = 3.8)
s_param = clf.fit(x ,y)
```

```python
# Plot fit

plt.scatter(x, y, marker='.', color='b')

x_fit = np.linspace(-3, 3, 100)
plt.plot(x_fit, 1 / (1 + np.exp(-(n_param[1] * x_fit + n_param[0]))), color='g', label='Newton')
plt.plot(x_fit, 1 / (1 + np.exp(-(b_param[1] * x_fit + b_param[0]))), color='r', label='Batch gradient')
plt.plot(x_fit, 1 / (1 + np.exp(-(s_param[1] * x_fit + s_param[0]))), color='k', label='Stochastic gradient')
plt.title('Fit Result')
plt.legend()

plt.show()
```

### Exponential Distribution

Let $Y \sim Exponential(\lambda)$, where the canonical link function is defined as $ \mu = X\beta^{-1}$

```python
# Simulate data with Exponential error distribution

x = np.linspace(1, 10, 100)
y = np.zeros(100)

mu = 1 / (1 + 2 * x)

for i in range(0, len(mu)):
    y[i] = np.random.exponential(mu[i])
```

```python
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('Probability')

plt.show()
```

```python
# Fit using py-glm

# Newton's method
clf = glm.GLM(distr='exponential', n_iter=1000, descent='N', learning_rate = .001)
n_param = clf.fit(x, y)

# Batch gradient descent
clf = glm.GLM(distr='exponential', n_iter=1000, descent='BGD', learning_rate = .0001)
b_param = clf.fit(x, y)

# Stochastic gradient descent
clf = glm.GLM(distr='exponential', n_iter=1000, descent='SGD', learning_rate = .001)
s_param = clf.fit(x, y)
```

```python
# Plot fit

plt.scatter(x, y, marker='.', color='b')
plt.plot(x, 1 / (n_param[1] * x + n_param[0]), color='g', label='Newton')
plt.plot(x, 1 / (b_param[1] * x + b_param[0]), color='r', label='Batch gradient')
plt.plot(x, 1 / (s_param[1] * x + s_param[0]), color='k', label='Stochastic gradient')
plt.title('Fit Result')
plt.legend()

plt.show()
```
